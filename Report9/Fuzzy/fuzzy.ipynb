# Import

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

# Fuzzification

Academic and Industrial reputation on subjective ranges [0, 10]
University reputation has a range of [0, 25] in units of percentage points

```python
x_ac = np.arange(0, 11, 1)
x_in = np.arange(0, 11, 1)
x_un  = np.arange(0, 26, 1)
```

Generate fuzzy membership functions

```python
ac_lo = fuzz.trimf(x_ac, [0, 0, 5])
ac_md = fuzz.trimf(x_ac, [0, 5, 10])
ac_hi = fuzz.trimf(x_ac, [5, 10, 10])
in_lo = fuzz.trimf(x_in, [0, 0, 5])
in_md = fuzz.trimf(x_in, [0, 5, 10])
in_hi = fuzz.trimf(x_in, [5, 10, 10])
un_lo = fuzz.trimf(x_un, [0, 0, 13])
un_md = fuzz.trimf(x_un, [0, 13, 25])
un_hi = fuzz.trimf(x_un, [13, 25, 25])
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 9))

ax0.plot(x_ac, ac_lo, 'b', linewidth=1.5, label='Low')
ax0.plot(x_ac, ac_md, 'g', linewidth=1.5, label='Medium')
ax0.plot(x_ac, ac_hi, 'r', linewidth=1.5, label='High')
ax0.set_title('Academic reputation')
ax0.legend()

ax1.plot(x_in, in_lo, 'b', linewidth=1.5, label='Low')
ax1.plot(x_in, in_md, 'g', linewidth=1.5, label='Medium')
ax1.plot(x_in, in_hi, 'r', linewidth=1.5, label='High')
ax1.set_title('Industrial reputation')
ax1.legend()

ax2.plot(x_un, un_lo, 'b', linewidth=1.5, label='Low')
ax2.plot(x_un, un_md, 'g', linewidth=1.5, label='Medium')
ax2.plot(x_un, un_hi, 'r', linewidth=1.5, label='High')
ax2.set_title('University reputation')
ax2.legend()


plt.tight_layout()
```

# Fuzzy rules

```python
ac_value = 3.5
in_value = 7.8

ac_level_lo = fuzz.interp_membership(x_ac, ac_lo, ac_value)
ac_level_md = fuzz.interp_membership(x_ac, ac_md, ac_value)
ac_level_hi = fuzz.interp_membership(x_ac, ac_hi, ac_value)

in_level_lo = fuzz.interp_membership(x_in, in_lo, in_value)
in_level_md = fuzz.interp_membership(x_in, in_md, in_value)
in_level_hi = fuzz.interp_membership(x_in, in_hi, in_value)
```

```python
# Bad Academic or Industrial reputation => Bad University reputation
active_rule1 = np.fmax(ac_level_lo, in_level_lo)
un_activation_lo = np.fmin(active_rule1, un_lo)

# Medium Academic reputation => Medium University reputation
un_activation_md = np.fmin(in_level_md, un_md)

# High Academic or Industrial reputation => High University reputation
active_rule3 = np.fmax(ac_level_hi, in_level_hi)
un_activation_hi = np.fmin(active_rule3, un_hi)
un0 = np.zeros_like(x_un)
```

```python
fig, ax0 = plt.subplots(figsize=(8, 3))

ax0.fill_between(x_un, un0, un_activation_lo, facecolor='b', alpha=0.7)
ax0.plot(x_un, un_lo, 'b', linewidth=0.5, linestyle='--', )
ax0.fill_between(x_un, un0, un_activation_md, facecolor='g', alpha=0.7)
ax0.plot(x_un, un_md, 'g', linewidth=0.5, linestyle='--')
ax0.fill_between(x_un, un0, un_activation_hi, facecolor='r', alpha=0.7)
ax0.plot(x_un, un_hi, 'r', linewidth=0.5, linestyle='--')
ax0.set_title('Output membership activity')

plt.tight_layout()
```

# Defuzzification

Aggregate all three output membership functions together

```python
aggregated = np.fmax(un_activation_lo,
                     np.fmax(un_activation_md, un_activation_hi))
```

```python
un_c = fuzz.defuzz(x_un, aggregated, 'centroid')
un_b = fuzz.defuzz(x_un, aggregated, 'bisector')
un_activation_c = fuzz.interp_membership(x_un, aggregated, un_c)
un_activation_b = fuzz.interp_membership(x_un, aggregated, un_b)
```

```python
fig, ax0 = plt.subplots(figsize=(8, 3))

ax0.plot(x_un, un_lo, 'b', linewidth=0.5, linestyle='--', )
ax0.plot(x_un, un_md, 'g', linewidth=0.5, linestyle='--')
ax0.plot(x_un, un_hi, 'r', linewidth=0.5, linestyle='--')
ax0.fill_between(x_un, un0, aggregated, facecolor='Orange', alpha=0.7)
ax0.plot([un_c, un_c], [0, un_activation_c], 'k', linewidth=1.5, alpha=0.9)
ax0.plot([un_b, un_b], [0, un_activation_b], 'c', linewidth=1.5, alpha=0.9)
ax0.set_title('Aggregated membership and result (line)')

plt.tight_layout()
```
